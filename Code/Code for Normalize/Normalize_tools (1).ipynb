```python
import pandas as pd
import numpy as np
from numpy import dstack
from sklearn.preprocessing import MinMaxScaler
from numpy import savetxt
from tensorflow.keras.utils import to_categorical
```

```python
# DataSet
dataset = pd.DataFrame({'Inorm_r':[],
           'Inorm_s':[],
           'Inorm_t':[],
           'Inorm_n':[],
           'Vnorm_r':[],
           'Vnorm_s':[],
           'Vnorm_t':[]})

# Banyaknya data tambah 1 karena dari 1
banyak_data = 101
data_Ir = np.array([])
data_Is = np.array([])
data_It = np.array([])
data_In = np.array([])
data_Vr = np.array([])
data_Vs = np.array([])
data_Vt = np.array([])

for i in range(1,banyak_data):
  
  #Set nama kolom
  columts = ['Data_ke', 'Freq', 'Ir','Is', 'It', 'In', 'Vr', 'Vs', 'Vt'] 

  # Baca data dari csv
  data = str(i) + ".csv"
  df  = pd.read_csv(data, names=columts)

  # Mencari nilai Min-Max I dan V
  column_i = df[['Ir','Is','It','In']]
  column_v = df[['Vr','Vs','Vt']]
  df['Imax'] = column_i.max(axis=1)
  df['Imin'] = column_i.min(axis=1)
  df['Vmax'] = column_v.max(axis=1)
  df['Vmin'] = column_v.min(axis=1)
  Imax = df['Imax'].max()
  Imin = df['Imin'].min()
  Vmax = df['Vmax'].max()
  Vmin = df['Vmin'].min()

  # Normalize data
  df['Inorm_r'] = df['Ir'].apply(lambda x: (x - Imin) / (Imax - Imin))
  df['Inorm_s'] = df['Is'].apply(lambda x: (x - Imin) / (Imax - Imin))
  df['Inorm_t'] = df['It'].apply(lambda x: (x - Imin) / (Imax - Imin))
  df['Inorm_n'] = df['In'].apply(lambda x: (x - Imin) / (Imax - Imin))
  df['Vnorm_r'] = df['Vr'].apply(lambda x: (x - Vmin) / (Vmax - Vmin))
  df['Vnorm_s'] = df['Vr'].apply(lambda x: (x - Vmin) / (Vmax - Vmin))
  df['Vnorm_t'] = df['Vr'].apply(lambda x: (x - Vmin) / (Vmax - Vmin))

  # Mengambil data Norm arus dan tegangan
  dfn = df[['Inorm_r','Inorm_s','Inorm_t','Inorm_n','Vnorm_r','Vnorm_s','Vnorm_t']]
  #print(dfn)

  # Mengambil data Norm kolom satu persatu
  Ir = dfn.loc[:,'Inorm_r']
  Is = dfn.loc[:,'Inorm_s']
  It = dfn.loc[:,'Inorm_t']
  In = dfn.loc[:,'Inorm_n']

  Vr = dfn.loc[:,'Vnorm_r']
  Vs = dfn.loc[:,'Vnorm_s']
  Vt = dfn.loc[:,'Vnorm_t']
  
  # Mengubahnya ke dalam bentuk numpy array
  Ir_temp = np.array(Ir.values)
  Is_temp = np.array(Is.values)
  It_temp = np.array(It.values)
  In_temp = np.array(In.values)
  
  Vr_temp = np.array(Vr.values)
  Vs_temp = np.array(Vs.values)
  Vt_temp = np.array(Vt.values)
  #print(Ir_temp.shape)

  # Menambahkan ke dalam data fitur
  data_Ir = np.append(data_Ir, Ir_temp)
  data_Is = np.append(data_Is, Is_temp)
  data_It = np.append(data_It, It_temp)
  data_In = np.append(data_In, In_temp)
  
  data_Vr = np.append(data_Vr, Vr_temp)
  data_Vs = np.append(data_Vs, Vs_temp)
  data_Vt = np.append(data_Vt, Vt_temp)
  #print(Ir_temp.shape, data_Ir.shape)


# Reshape data ke 2D
print(data_Ir.shape, data_Is.shape, data_It.shape, data_In.shape, data_Vr.shape, data_Vs.shape, data_Vt.shape)
data_Ir = data_Ir.reshape(banyak_data-1, 255)
data_Is = data_Is.reshape(banyak_data-1, 255)
data_It = data_It.reshape(banyak_data-1, 255)
data_In = data_In.reshape(banyak_data-1, 255)
  
data_Vr = data_Vr.reshape(banyak_data-1, 255)
data_Vs = data_Vs.reshape(banyak_data-1, 255)
data_Vt = data_Vt.reshape(banyak_data-1, 255)

print(data_Ir.shape, data_Is.shape, data_It.shape, data_In.shape, data_Vr.shape, data_Vs.shape, data_Vt.shape)

print("mantap mamang")

# Ekspor ke CSV
savetxt('data Ir.csv', data_Ir, delimiter=',')
savetxt('data Is.csv', data_Is, delimiter=',')
savetxt('data It.csv', data_It, delimiter=',')
savetxt('data In.csv', data_In, delimiter=',')
savetxt('data Vr.csv', data_Vr, delimiter=',')
savetxt('data Vs.csv', data_Vs, delimiter=',')
savetxt('data Vt.csv', data_Vt, delimiter=',')
```

```python
# Membuat Label
# 1 - 42 = Hubung Singkat (1)
HS = 42
# 43 - 67 = Non-System Fault (2)
NSF = 25
# 68 - 100 = Petir (3)
PTR = 33

label = np.array([])

for data in range(0, HS):
  label = np.append(label, [1])
for data in range(0, NSF):
  label = np.append(label, [2])
for data in range(0, PTR):
  label = np.append(label, [3])

# Ubah data label ke bentuk binner
label = to_categorical(label)   

# Ekspor data label
savetxt('label.csv', label, delimiter=',')
```
